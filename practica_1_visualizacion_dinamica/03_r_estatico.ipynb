```python
library(sf)
library(tmap)
library(ggplot2)

data("World")

col <- World[World$name == "Colombia", ]

tm_shape(col) +
  tm_polygons("pop_est", palette = "Blues", title = "Población Est.") +
  tm_compass(position = c("right", "top")) +
  tm_scale_bar()

ggplot() +
  geom_sf(data = col, fill = "orange", color = "black") +
  theme_minimal() +
  labs(title = "Mapa de Colombia con ggplot2")
```
